# Su Kalitesi (İçilebilir Su) Veri Seti Kullanarak Veri Analizi


# Veri Seti Hakkında

Bu veri seti analizi, suyun içilebilirliğini belirlemek için çeşitli parametreleri kullanmayı amaçlamaktadır. pH, sertlik, toplam çözünmüş katılar (TDS), kloraminler, sülfat, iletkenlik ve diğer önemli özellikler üzerinde yapılan analizler, suyun insan tüketimi için güvenli olup olmadığını değerlendirmeyi hedeflemektedir.  Veri seti Kaggle platformundan (https://www.kaggle.com/datasets/adityakadiwal/water-potability)  alınmıştır.

Veri setinin içindeki niteliklerin tanımları aşağıdaki gibidir.


* __ph (pH değeri):__ pH, suyun asit-baz dengesini değerlendirmede önemli bir parametredir. Ayrıca, suyun asidik veya alkalin durumunu gösteren bir belirteçtir. Dünya Sağlık Örgütü (WHO), pH için maksimum izin verilen sınırı 6,5 ile 8,5 arasında önermiştir. Mevcut araştırma aralığı 6,52 ile 6,83 arasındadır ve bu da WHO standartlarına uygundur.

* __Hardness (Sertlik):__ Sertlik genellikle kalsiyum ve magnezyum tuzlarından kaynaklanır. Bu tuzlar, suyun geçtiği jeolojik tabakalardan çözünmüştür. Su, sertlik oluşturan malzemelerle temas süresinin uzunluğu, ham su içinde ne kadar sertlik olduğunu belirlemeye yardımcı olur. Sertlik, başlangıçta, Kalsiyum ve Magnezyum tarafından meydana getirilen sabunun çökmesi kapasitesi olarak tanımlanmıştır.

* __Solids (Katılar (Toplam çözünmüş katılar - TDS)):__ Su, potasyum, kalsiyum, sodyum, bikarbonatlar, klorürler, magnezyum, sülfatlar vb. gibi geniş bir inorganik ve bazı organik mineralleri veya tuzları çözebilme özelliğine sahiptir. Bu mineraller, suda istenmeyen tat ve seyreltilmiş renk oluşumuna neden olur. Bu, suyun kullanımı için önemli bir parametredir. Yüksek TDS değerine sahip su, suyun yüksek mineral içeriğini gösterir. TDS için istenen limit 500 mg/l'dir ve maksimum limit içme amacıyla önerilen 1000 mg/l'dir.

* __Chloramines (Kloraminler):__ Klorin ve kloramin, kamu su sistemlerinde kullanılan başlıca dezenfektanlardır. Kloraminler, içme suyunu arıtmak için kloraminin amonyak ile birleştirilmesiyle en yaygın olarak oluşturulur. İçme suyunda, klorin seviyeleri 4 miligram/litre (mg/L veya 4 parçaya milyon (ppm)) kadar güvenli kabul edilir.

* __Sulfate (Sülfat):__ Sülfatlar, mineral, toprak ve kayalarda bulunan doğal maddelerdir. Havada, yeraltı suyunda, bitkilerde ve yiyeceklerde bulunurlar. Sülfatın başlıca ticari kullanımı kimya endüstrisindedir. Deniz suyundaki sülfat konsantrasyonu yaklaşık 2.700 miligram/litre (mg/L) dir. Tatlı su kaynaklarının çoğunda 3 ila 30 mg/L arasında değişir, ancak bazı coğrafi konumlarda (1000 mg/L) çok daha yüksek konsantrasyonlar bulunur.

* __Conductivity (İletkenlik):__ Saf su, elektrik akımını iletme yeteneğine sahip değildir, daha ziyade iyi bir yalıtkandır. İyon konsantrasyonundaki artış, suyun elektriksel iletkenliğini artırır. Genellikle, suya çözünmüş katıların miktarı, elektriksel iletkenliği belirler. Elektrik iletkenliği (EC) aslında bir çözeltinin iyonik sürecini ölçer ve bu, akımı iletebilmesini sağlar. Dünya Sağlık Örgütü standartlarına göre, EC değeri 400 μS/cm'yi aşmamalıdır.

* __Organic_carbon (Organik karbon):__ Toplam Organik Karbon (TOC), kaynak sularında çürüyen doğal organik maddenin (NOM) yanı sıra sentetik kaynaklardan da gelir. TOC, saf su içindeki organik bileşiklerdeki toplam karbon miktarını ölçer. ABD Çevre Koruma Ajansı'na göre, işlenmiş / içme suyunda TOC için <2 mg/L, işlem için kullanılan kaynak suyunda ise <4 mg/Lit olmalıdır.

* __Trihalomethanes(Trihalometanlar):__ THM'ler, klor ile arıtılmış su içinde bulunabilen kimyasallardır. İçme suyundaki THM konsantrasyonu, suyun içindeki organik malzemenin seviyesine, suyu arıtmak için gereken klor miktarına ve arıtılan suyun sıcaklığına bağlı olarak değişir. İçme suyunda THM seviyeleri 80 ppm'ye kadar güvenli kabul edilir.

* __Turbidity (Bulanıklık):__ Su bulanıklığı, askıdaki katı maddenin miktarına bağlıdır. Bu, suyun ışık yayan özelliklerinin bir ölçüsüdür ve test, kolloidal madde açısından atık deşarjının kalitesini göstermek için kullanılır. Wondo Genet Kampüsü için elde edilen ortalama bulanıklık değeri (0,98 NTU), WHO tarafından önerilen 5,00 NTU değerinin altındadır.

* __Potability (İçilebilirlik):__ 1 insan tüketimi için Güvenli olanı ve 0 insan tüketimi için Güvensiz olanı gösterir.

## Motivasyon ve Arka Plan
### Seçilen Problemin Açıklaması ve Önemi
Su, yaşamın sürdürülebilmesi için en temel gereksinimlerden biridir. İçilebilir su kaynaklarının kalitesi, halk sağlığı, tarım, sanayi ve ekosistemler üzerinde doğrudan etkilere sahiptir. Günümüzde dünya genelinde su kaynaklarının kirlenmesi ve sürdürülebilir yönetimi, kritik bir çevresel ve toplumsal sorundur. Bu bağlamda, içilebilir su kalitesinin değerlendirilmesi ve izlenmesi, kamu sağlığını koruma ve çevre yönetimi açısından büyük önem taşımaktadır.
İçme suyu kalitesinin belirlenmesi ve izlenmesi, halk sağlığı açısından doğrudan sonuçlar doğurur. Kirli su tüketimi, çeşitli hastalıklara ve sağlık sorunlarına yol açabilir. Bu nedenle, suyun içilebilir olup olmadığının bilinmesi, sağlık risklerinin azaltılması ve halkın güvenli suya erişiminin sağlanması açısından kritik öneme sahiptir. Ayrıca, su kalitesinin sürdürülebilir yönetimi, doğal ekosistemlerin korunması ve su kaynaklarının gelecek nesillere aktarılması için de gereklidir.

# Araştırma sorularına ve sonuçlarına genel bakış


### 1- Değişken arasındaki ilişkileri yorumlayınız.

Nitelikler arasındaki ilişkileri yorumlayabilmek için korelasyon matrisi kullanılmıştır.Korelasyon katsayısı, iki değişken arasındaki doğrusal ilişkiyi ölçer ve -1 ile +1 arasında değişir. +1 mükemmel pozitif korelasyonu, -1 mükemmel negatif korelasyonu ve 0 korelasyon olmadığını gösterir. Matrisin renk skalası, daha koyu kırmızının daha yüksek pozitif korelasyonu ve daha koyu mavinin daha yüksek negatif korelasyonu gösterdiği bir gradyan sunar.

* pH ile Diğer Değişkenler: pH ile diğer değişkenler arasında belirgin bir korelasyon gözlenmemektedir. En yüksek korelasyon 0.089 ile Solids arasında olup, bu da oldukça düşük bir korelasyon anlamına gelir.

* Hardness ile Diğer Değişkenler:  Hardness ile diğer değişkenler arasında da anlamlı bir korelasyon yoktur. Örneğin, Sulfate (sülfat) ile -0.11 ve pH ile 0.082 arasında düşük korelasyon değerleri gözlemlenmektedir.

* Solids ile Diğer Değişkenler:  Solids ile Sulfate arasında -0.17'lik bir korelasyon var, bu da zayıf negatif bir ilişki olduğunu gösterir. Ancak, genellikle düşük korelasyonlar vardır.

* Chloramines ile Diğer Değişkenler: Chloramines ile diğer değişkenler arasında da belirgin bir korelasyon gözlenmemektedir. En yüksek korelasyon, Sulfate (sülfat) ile 0.027 ve bu değer çok düşük pozitif korelasyon anlamına gelir.

* Sulfate ile Diğer Değişkenler: Sulfate ile diğer değişkenler arasında belirgin bir korelasyon gözlenmemektedir. En yüksek korelasyon 0.031 ile Organic_carbon (organik karbon) arasında olup, bu da oldukça düşük bir korelasyon anlamına gelir.

* Conductivity ile Diğer Değişkenler: Conductivity ile diğer değişkenler arasında da belirgin bir korelasyon yoktur. En yüksek korelasyon, Organic_carbon (organik karbon) ile 0.021'dir.
* Organic_carbon ile Diğer Değişkenler: Organic_carbon ile diğer değişkenler arasında belirgin bir korelasyon gözlenmemektedir. En yüksek korelasyon 0.044 ile pH arasında olup, bu da oldukça düşük bir korelasyon anlamına gelir.

* Trihalomethanes ile Diğer Değişkenler: Trihalomethanes ile diğer değişkenler arasında belirgin bir korelasyon gözlenmemektedir. En yüksek korelasyon 0.017 ile Chloramines (kloraminler) arasında olup, bu da oldukça düşük bir korelasyon anlamına gelir.

* Turbidity ile Diğer Değişkenler: Turbidity ile diğer değişkenler arasında belirgin bir korelasyon gözlenmemektedir. En yüksek korelasyon 0.02 ile Solids (katı maddeler) arasında olup, bu da oldukça düşük bir korelasyon anlamına gelir.

* Genel Değerlendirme: Bu matris genel olarak düşük korelasyonlar göstermektedir, bu da değişkenler arasında güçlü doğrusal ilişkilerin olmadığını işaret eder. Her bir değişkenin diğerlerinden bağımsız olarak değerlendirilmesi gerektiğini söyleyebiliriz. Genelde, korelasyon değerleri 0.1'in altında kalmakta ve bu da bu değişkenlerin birbirleriyle önemli bir doğrusal ilişki göstermediğini ortaya koymaktadır.




### 2- Veri Setindeki Aykırı Değerler Her Zaman Düzenlenmeli mi ?

Aykırı veriler, genellikle veri setindeki diğer gözlemlerden önemli ölçüde farklı olan verilerdir. Ancak, bu verilerin her zaman hatalı veya anlamsız olduğunu söylemek doğru olmaz. Özellikle su kalitesi gibi bir konuda, belirli ölçümler (örneğin pH) aykırı görünebilir ama aslında anlamlı ve önemli bilgiler taşıyabilir. Örneğin pH değerinin 1 veya 13 olması, suyun içilemez olduğunu gösteren anlamlı bir veri olabilir. Bu tür veriler, suyun durumu hakkında kritik bilgiler sağlayabilir. Bu durumu göz önüne aldığımızda, Box-Plot tekniği kullandığımızda ortaya çıkan aykırı veriler aslında suyun içilebilirliği hakkında önemli bilgiler sunuyor olabilir. Tüm bunlardan yola çıkarak, model eğitiminin doğruluğunun daha yüksek olması için bu projede aykırı veriler sadece gösterilmiş olup, veri seti üzerinde herhangi bir aykırı veri giderme işlemi gerçekleştirilmemiştir.


### 3- Bu Veri Seti Hangi Türdedir ? Hangi Algoritmalar için uygundur ?

Bu veri seti, su kalitesini değerlendirmek için çeşitli nitelikleri içeren ve tamamen nümerik verilere dayanan bir veri setidir. Nitelikler arasında pH, sertlik (Hardness), katı maddeler (Solids), kloraminler (Chloramines), sülfat (Sulfate), iletkenlik (Conductivity), organik karbon (Organic_carbon), trihalometanlar (Trihalomethanes), bulanıklık (Turbidity) ve suyun içilebilirlik durumu (Potability) gibi değişkenler bulunmaktadır. Bu veri seti, suyun içilebilirliğini belirlemeye yönelik bir sınıflandırma problemi olarak ele alınabilir. Bu nedenle, lojistik regresyon, karar ağaçları, rastgele ormanlar, destek vektör makineleri, KNN ve yapay sinir ağları gibi denetimli öğrenme algoritmaları için uygundur. Ayrıca, K-Ortalamalar kümeleme ve PCA gibi denetimsiz öğrenme ve boyut azaltma teknikleri de veri setinin analizi ve görselleştirilmesi için kullanılabilir. Bu algoritmaların doğru uygulanması, su kalitesinin doğru bir şekilde tahmin edilmesine ve değerlendirilmesine yardımcı olacaktır.

* Lojistik Regresyon (Logistic Regression): Suyun içilebilir olup olmadığını ikili sınıflandırma yaparak tahmin etmek için kullanılabilir.
* Karar Ağaçları (Decision Trees): Karar ağaçları, nümerik ve kategorik verilerle çalışabilir ve değişkenler arasındaki ilişkiyi görselleştirerek su kalitesini değerlendirmede etkili olabilir.
* Random Forests: Birden fazla karar ağacını kullanarak daha yüksek doğruluk ve genelleme yeteneği sağlar. Su kalitesini tahmin etmek için oldukça etkili bir yöntemdir.
* Destek Vektör Makineleri (Support Vector Machines, SVM): Sınıflandırma problemlerinde güçlü performans gösterir ve su kalitesi verilerini sınıflandırmak için uygundur.
* K-En Yakın Komşu (K-Nearest Neighbors, KNN): Basit ama etkili bir sınıflandırma algoritmasıdır. Su kalitesi verilerini gruplamak ve sınıflandırmak için kullanılabilir.
* Yapay Sinir Ağları (Artificial Neural Networks): Özellikle karmaşık ilişkileri öğrenmek ve yüksek doğruluk elde etmek için kullanılabilir.

Biz bu projede KNN, Decision Tree, Random Forest algoritmalarını kullandık.

### 4- Bu Veri Seti İçin Normalizasyon Yapılması Şart Mıdır ?

Her veri seti için normalizasyon yapmak gerekli değildir ancak bazı durumlarda normalizasyon, model performansını önemli ölçüde artırabilir.Normalizasyon, veri setindeki özelliklerin belirli bir aralığa (genellikle 0 ile 1 arasında) ölçeklenmesi işlemidir. Bu durum, özellikle mesafe tabanlı ve gradyan tabanlı algoritmaların performansını olumsuz etkileyebilirken veri bölme kriterlerini (örneğin, Gini impurity veya bilgi kazancı) kullanarak çalışan algoritmaları genellikle olumsuz etkilemez. Biz projemizde KNN algoritmasını kullandığımız için ve algoritmanın yanlı öğrenmesini, aşırı öğrenmesini engellemek için normalizasyon yapmaya karar verdik.

## Normalizasyon

Veri normalleştirme, bir veri kümesinin değerlerini ortak bir ölçeğe dönüştürmek için kullanılan bir tekniktir. Birçok makine öğrenimesi algoritması,  girdi niteliklerinin ölçeğine duyarlıdır ve veriler normalleştirildiğinde daha iyi sonuçlar elde edilebilir. Basit bir ifadeyle, birden fazla öznitelik mevcut olduğunda ancak özniteliklerin farklı ölçeklerde değerleri varsa, bu, veri analizi işlemleri gerçekleştirilirken zayıf veri modellerine yol açabilir[2].Literatürde birden fazla normalizasyon teknikleri bulunmaktadır bunlardan bazıları: 
* __Min-Maks normalizasyonu:__ Bu teknik, bir özelliğin değerlerini 0 ile 1 arasında bir aralığa ölçeklendirir. Bu işlem, özelliğin minimum değerinin her bir değerden çıkarılması ve ardından özelliğin aralığına bölünmesiyle gerçekleştirilir.

* __Z-puanı normalizasyonu:__ Bu teknik, bir özelliğin değerlerini ortalama 0 ve standart sapma 1 olacak şekilde ölçeklendirir. Bu işlem, özelliğin ortalamasının her değerden çıkarılması ve ardından standart sapmaya bölünmesiyle gerçekleştirilir.

* __Ondalık Ölçeklendirme:__ Bu teknik, bir özelliğin değerlerini 10'un katlarına bölerek bir özelliğin değerlerini ölçeklendirir.

* __Logaritmik dönüşüm:__ Bu teknik, bir özelliğin değerlerine logaritmik dönüşüm uygular. Bu işlem, aykırı değerlerin etkisini azaltmaya yardımcı olabileceğinden geniş bir değer aralığına sahip veriler için yararlı olabilir.

* __Kök dönüşümü:__ Bu teknik, bir özelliğin değerlerine karekök dönüşümü uygular. Bu işlem, aykırı değerlerin etkisini azaltmaya yardımcı olabileceğinden geniş bir değer aralığına sahip veriler için yararlı olabilir.

Bu proje için Min-Maks normalizasyonu tercih edilerek veri setinde bulunan tüm veriler 0 ile 1 arasında ölçeklendirilmiştir. Aşağıdaki formül kullanılarak Min-Maks normalizasyon işlemi gerçekleştirimiştir.

## Aykırı Değer

Diğer değerlerle karşılaştırıldığında veri setine uygun olmadığı tespit edilen aşırı değerlere aykırı değer denir. Aykırı değerlerin fazla olması veri setinin normal dağılımdan sapmasına ve yapacağımız istatiksel analiz ve uygulayacağımız algoritmaların yanılmasına sebep olabilirler[https://uzaktanegitim23bahar.erbakan.edu.tr/pluginfile.php/791738/mod_folder/content/0/Ayk__r__%20De__er%20Y__netimi%5B%23207962%5D-182758.pdf?forcedownload=1]. Bu nedenle aykırı değerler tespit edilmeli, düzenlenmeli ve gerekirse veri setinden çıkartılmalıdır. Aykırı değerleri tespit ederken kullanılan yöntemlerden bazıları;

 **Görsel Yöntemler:** Veri setini görselleştirerek aykırı değerleri belirleme. Box plot (kutu grafiği), scatter plot (dağılım grafiği), histogramlar ve qq plot (quantile-quantile plot) gibi grafikler kullanılabilir.

**İstatistiksel Yöntemler:**

- **Z-Skoru veya Standart Sapma Yaklaşımı:** Veri setinin ortalama ve standart sapmasını kullanarak, belirli bir eşik değerinden (genellikle 2 veya 3 standart sapma) daha fazla uzaklıkta olan değerleri aykırı olarak tanımlama.

- **IQR (Interquartile Range) Yaklaşımı:** Verileri çeyrekler arası aralıklara (Q1 ve Q3) bölen ve bunlar arasındaki farkı (IQR) kullanarak aykırı değerleri tanımlama. Genellikle Q1 - 1.5 * IQR ve Q3 + 1.5 * IQR sınırları kullanılır.

**Makine Öğrenimi Yöntemleri:** Bir veri setindeki aykırı değerleri belirlemek için özel olarak tasarlanmış algoritmalar kullanma. Bu algoritmalar arasında Isolation Forest, Local Outlier Factor (LOF), One-Class SVM (Support Vector Machine) gibi yöntemler bulunur.

Bu aşamada, aykırı değerlerin belirlenmesinde Box plot grafiği kullanılmıştır. İlk aşamada çizilen Box plot grafiği, aykırı verilerin mevcudiyetini göstermiştir. Bununla birlikte, her bir nitelikte bulunan aykırı veriler,alt sınır veya üst sınır değerlerine eşitlenmesi yöntemi kullanılarak yapılan düzenleme sonrasında çizilen Box plot grafiği incelendiğinde, aykırı değerlerin giderildiği gözlenmiştir. Bu yöntem, veri setindeki aykırı değerlerin etkisini azaltarak analiz sonuçlarının daha güvenilir olmasını sağlamıştır.

## Eksik Veri Tamamlama[3]
Eksik değerler, bir veri kümesinde belirli bir nitelikte bulunmayan veri noktalarıdır. Bunlar, "NA" veya "bilinmeyen" gibi özel semboller gibi çeşitli yollarla temsil edilebilir. Bu eksik veri noktaları, veri analizinde önemli bir zorluk oluşturarak yanlış veya önyargılı sonuçlara yol açabilir.

* Örneklem büyüklüğünü azaltabilir: Bu, analizinizin doğruluğunu ve güvenilirliğini azaltabilir.

* Önyargıya sebep olabilirlir: Eksik veriler uygun şekilde ele alınmazsa yanlış bir analiz sonuucuna sebebiyet verebilir.

* Belirli analizleri yapmayı zorlaştırabilir: Bazı istatistiksel teknikler, eksik değerler bulunduğunda tüm değişkenler için tam veri gerektirir, bu da bu tekniklerin uygulanamaz hale gelmesine neden olabilir.

### Eksik Veri Türleri

Üç ana eksik veri türü bulunmakadır:

* __Tamamen Rastgele Eksik (MCAR):__ MCAR, bir veri noktasının eksik olma olasılığının tamamen rastgele ve veri kümesindeki herhangi başka bir değişkenden bağımsız olduğu belirli bir eksik veri türüdür. Daha basit bir ifadeyle, bir değerin eksik olup olmadığı, diğer değişkenlerin değerleri veya veri noktasının özellikleriyle hiçbir ilgisi yoktur.

* __Rastgele Olarak Eksik (MAR):__ MAR, bir veri noktasının eksik olma olasılığının veri kümesindeki diğer değişkenlerin değerlerine bağlı olduğu, ancak eksik değişkenin kendisine bağlı olmadığı bir eksik veri türüdür. Bu, eksiklik mekanizmasının tamamen rastgele olmadığı, ancak mevcut bilgilere dayanarak tahmin edilebileceği anlamına gelir.

* __Rastgele Olmayan Eksik (MNAR):__ MNAR, başa çıkılması en zor eksik veri türüdür. Bir veri noktasının eksik olma olasılığının eksik değerle ilişkili olduğu durumdur. Bu, eksik verinin nedeninin bilgilendirici olduğu ve eksik olan değişkenle doğrudan ilişkili olduğu anlamına gelir.

### Eksik Verilerin Giderilme Yöntemleri

* __Eksik Verileri Silme:__ Eğer eksik verilerin oranı çok düşükse ve analizinize ciddi bir etkisi olmayacaksa, eksik değerlere sahip gözlemleri silmek bir seçenek olabilir. Ancak, bu yöntem veri kaybına neden olabilir ve analizinizi yanıltabilir.

* __Değer Atama__: Eksik değerleri bir değerle doldurmak, eksik veri problemini çözmek için sıkça kullanılan bir yöntemdir. Bu değeri seçerken, eksik verilerin dağılımına, değişkenin özelliklerine ve veri setinizin yapısına bağlı olarak ortalama, medyan, en yakın komşu, regresyon veya doldurma yöntemlerini kullanabilinir.

* __Yapay Zeka Teknikleriri:__ Eksik verileri tahmin etmek için makine öğrenimi modelleri kullanabilirsiniz. Örneğin, eksik değerleri tahmin etmek için regresyon veya sınıflandırma modelleri oluşturabilinir.

Bu yaklaşım, algoritmanın farklı kalıplara uyum sağlamasına ve verilerin yerel yapısına göre tahminlerde bulunmasına olanak tanır [7]. Çoğunluk oyu terimi, genellikle belirli bir veri noktası çevresinde en sık temsil edilen sınıf etiketinin kullanılmasını ifade eder. Örneğin, KNN  algoritması bir veri noktasını sınıflandırırken, çevresindeki en yakın K komşudan elde edilen  etiketler arasında çoğunluk oyu prensibini kullanarak yeni veri noktasına bir sınıf etiketi atar.

## Karar Ağacı

Karar Ağaçları, sınıflandırma ve regresyon için kullanılan parametrik olmayan denetimli bir öğrenme yöntemidir. Karar ağaçları, her düğümde veri kümesini bölerek ve belirli bir ölçüt kullanarak en homojen alt grupları elde ederek işler. Bu bölünmeler, her düğümde en iyi ayrımı sağlayan özellikleri seçmek için yapılan kararlarla gerçekleşir. Sonuç olarak, ağaç, kökten başlayarak yapraklara doğru genişler ve her yaprakta bir hedef değişken değeri tahmin edilir.

* __Modelin Karmaşıklığı:__ Büyük ve karmaşık veri setleriyle çalışırken, karar ağaçları aşırı karmaşık hale gelebilir ve bu da modelin anlaşılmasını zorlaştırabilir.

* __Dengesiz Veri Setleriyle Başa Çıkma:__ Dengesiz sınıf dağılımlarına sahip veri setleriyle çalışırken, karar ağaçlarının doğruluğu azalabilir. Bu nedenle, dengesiz veri setlerini dengelemek için ön işleme adımları gerekebilir.

# Modellerin Karşılaştırılması 

Kruskal-Wallis testi parametrik olmayan bir testtir ve Tek Yönlü Anova'ya bir alternatiftir. Parametrik olmayan derken, verilerin belirli bir dağılımdan geldiği varsayılmamaktadır. Bu testin ana amacı, en az üç bağımsız grubun medyanları arasında istatistiksel bir fark olup olmadığını belirlemektir.Kruskal-Wallis Testi aşağıda tartışıldığı gibi, sıfır ve alternatif hipotezlere sahiptir [11]:

Sıfır hipotezi (H0): Tüm veri grupları için medyan aynıdır.
Alternatif hipotez (Ha): Tüm veri grupları için medyan eşit değildir.

Bu projede modellerin karşılaştırılması için Kruskal- Wallis testi tercih edilmiştir. Sonuçları incelediğimizde gruplar arasında herhangi bir fark olmadığını görürüz. P-değeri küçüldükçe, gruplar arasında anlamlı bir fark olduğu sonucuna varırız. Bu durumda, p-değeri 0.05 anlamlılık düzeyini (genellikle kabul edilen standart bir anlamlılık düzeyi) geçmediği için, null hipotezi reddetmek için yeterli kanıt yoktur. Yani, test sonuçlarına göre, kullanılan üç algoritma arasında istatistiksel olarak anlamlı bir farklılık bulunmamaktadır.

```python
# KNN
knn_metrics = {
    'accuracy': 0.6399,
    'precision': 0.60,
    'recall': 0.22,
    'f1_score': 0.32
}

# Decision Tree
dt_metrics = {
    'accuracy': 0.57,
    'precision': 0.45,
    'recall': 0.41,
    'f1_score': 0.43
}

# Random Forest
rf_metrics = {
    'accuracy': 0.65,
    'precision': 0.60,
    'recall': 0.30,
    'f1_score': 0.40
}

# Metrikler listesi
metric_lists = [list(knn_metrics.values()), list(dt_metrics.values()), list(rf_metrics.values())]

# Kruskal-Wallis testini uygula
_, p_value = kruskal(*metric_lists)

# Anlamlılık seviyesini belirle
alpha = 0.05

# Sonuçları kontrol et
if p_value < alpha:
    print("En az bir model diğerlerinden istatistiksel olarak anlamlı derecede farklıdır.")
    # En iyi modeli belirle
    best_model_index = metric_lists.index(max(metric_lists))
    if best_model_index == 0:
        print("KNN en iyi modeldir.")
    elif best_model_index == 1:
        print("Decision Tree en iyi modeldir.")
    else:
        print("Random Forest en iyi modeldir.")
else:
    print("Modeller arasında istatistiksel olarak anlamlı bir fark yoktur. Model seçimi için daha fazla inceleme yapılabilir.")
```

Confusion matrisine göre modelin performansı şu şekilde değerlendirilebilir:

### Confusion Matrisinin Yorumu

- **True Negatives (TN)**: 545
  - Model, 545 örneği doğru bir şekilde içilemeyen su olarak sınıflandırmış.
- **False Positives (FP)**: 55
  - Model, 55 içilemeyen su örneğini yanlış bir şekilde içilebilir olarak sınıflandırmış.
- **False Negatives (FN)**: 299
  - Model, 299 içilebilir su örneğini yanlış bir şekilde içilemeyen olarak sınıflandırmış.
- **True Positives (TP)**: 84
  - Model, 84 örneği doğru bir şekilde içilebilir su olarak sınıflandırmış.

### Performans Değerlendirmesi

1. **Doğru Pozitif Oranı (True Positive Rate / Recall)**:
   - İçilebilir suyu doğru tespit etme oranı: {TP} / {TP + FN} = {84} / {84 + 299} ≈ 0.22
   - Model, içilebilir suyu tespit etmede düşük bir başarı gösteriyor.
<br></br>
2. **Doğru Negatif Oranı (True Negative Rate / Specificity)**:
   - İçilemeyen suyu doğru tespit etme oranı: {TN} / {TN + FP} = {545} / {545 + 55} ≈ 0.9 
   - Model, içilemeyen suyu tespit etmede orta derecede başarılı.
<br></br>
3. **Pozitif Tahmin Değeri (Precision)**:
   - İçilebilir su olarak sınıflandırılan verinin gerçekten içilebilir olma oranı: {TP} / {TP+FP} = {84} / {84 + 55} ≈ 0.6 
<br></br>
4. **Genel Doğruluk Ay (Accuarcy)**:
   - Modelin genel doğruluğu: {TP + TN} / {TP + TN + FP + FN} ≈ {84+ 545} / {84 + 545 + 55 + 299} = 0.64 
<br></br>   
5. **F1 Score**:
   - Modelin genel doğruluğu: {2 * Prescision * Recall} / {Presicion + Recall} ≈ {2 * 0.6 * 0.22} / { 0.6 + 0.2} = 0.33
 


### Genel Değerlendirme

- **Yanlış Pozitifler (False Positives - FP)**: Model, içilemeyen 55 örneği içilebilir olarak sınıflandırmış. Bu, halk sağlığı açısından önemli bir risk oluşturabilir çünkü insanlar bu suyu güvenli sanarak tüketebilir.
- **Yanlış Negatifler (False Negatives - FN)**: Model, içilebilir 299 örneği içilemeyen olarak sınıflandırmış. Bu, kaynakların gereksiz yere reddedilmesine ve su israfına yol açabilir.
- **Doğru Negatifler (True Negatives - TN)** ve **Doğru Pozitifler (True Positives - TP)** sayıları, modelin bazı durumlarda doğru çalıştığını ancak içilebilir suyu tespit etmede daha zayıf olduğunu göstermektedir.
